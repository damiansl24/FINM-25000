# Order types

market order -> no spec price (execute whatever price is available when it reaches the floor)

Limit order -> wish to buy or sell at a spec price. Executed at spec price or better
- buy limit (below current mkt price)
- Sell limit (above mkt limit orders may not be executed if there is stock ahead (orders at same price with higher priority))

Stop order -> becomes mkt order once specified price is reached / passed. Stop price.
- Sell (below current mkt price)
- Buy (above current mkt price. Limit a loss or protect a profit ona short sale)

Stop limit order -> stop price activates order: once activated, becomes buy or sell limit order and can only be executed at a specific price or better
- risk: order may not get filled.
- sell stop limit -> always below mkt price
- But stop limit order: above mkt price.

# Exchange and trading systems

In the exchange there are participants. Many people want to trade (buy/sell) and so the exchange needs to broadcast to all market participants that an order has been made on the exchange.

Orders sit on the order book in the exchange.

With the order book (containing a list of limit orders). When there is someone interested in buying / selling, exchange will update the order books. If there is someone who matches the order, the exchange will be in charge of matching them.

## matching engine algorithm

Exchange receives an order: checks if the order can be matched with the order book.

Inputs: order and current order book.

Output of the algorithm is the modified order book and the trade that was executed. 

### Example:

Buy order comes in: existing sell orders on the exchange. Buy order matches with the best price on the order book currently. 

I want to buy 1 share @ $100, and there are two sell orders in the book: sell 1 @$99 and sell 1 @ $100. The trade matched is then 1 @ $99, and then the order book will then keep the sell of 1 share @ $100. 

(Best price offer model)

### Matching engine: partially filled order. 

buy 4@100: sell 1@99, sell 2@100. Engine matches the first best price (1@99 and 2@100). What remains in the order book is the buy 1@100 (unresolved order). When you take liquidity from the market (get 1  @99 and you wanted 1 @ 100) the exchange takes the surplus, but if you get a discount the exchange gives you a credit. 

### What happens if there are two sell orders that could fill the buy order?

FIFO -> first one who put in the sell order will be matched. Problem: fastest always gets everything, discouraging liquidity in the market. 

Uses pro rata.
- Market participants entering order is shared evenly among matching counter orders proportionally to their magnitude (pure pro rata)
- modified pro rata (combination of best price and pro rata)
