```python
import yfinance as yf
import sklearn as sk
import numpy
```

```python
goog = yf.download("GOOG", start = "2008-01-01", end = "2018-01-01")
```

```python
win = 20
slope_list = []
for i in range(len(goog)):
    if i < win:
        slope_list.append(None)
    else:
        x = numpy.arange(win)
        y = goog['Close'].iloc[i-win:i]
        slope, intercept = numpy.polyfit(x, y, 1)
        slope_list.append(slope)

goog['slope'] = slope_list
```

```python
signal_list = []
for slope in goog['slope']:
    if slope == None:
        signal_list.append(0)
    elif slope > 0:
        signal_list.append(1)
    else:
        signal_list.append(-1)

goog['signal'] = signal_list
```

```python
goog['returns'] = goog['Close'].pct_change()
goog['strategy'] = goog['signal'].shift(1) * goog['returns']
```

```python
strategy_return = goog['strategy'].cumsum().iloc[-1]
```

```python
print(strategy_return)
```
